```python
import requests

# 1. The Endpoint
# This is the exact address of the resource we want.
url = "https://jsonplaceholder.typicode.com/users/1"

# 2. The Execution
# Python sends the packet across the internet and waits (blocks) 
# until the server replies.
response = requests.get(url)

# 3. The Response Object
# 'response' is NOT the data. It is an object containing the data, headers, and time.
print(f"Server replied with: {response.status_code}")

# 4. Extracting Data
# .text gives us the raw string (useful for HTML)
# .json() parses the string into a Python Dictionary (CRITICAL STEP)
if response.status_code == 200:
    data = response.json()
    
    # Now we use standard Python Dictionary syntax
    print(f"Name: {data['name']}")
    print(f"City: {data['address']['city']}") # Nested access
    print(f"Company: {data['company']['name']}")
else:
    print("Request failed.")
```

```python
import requests

# 1. Use the real JSONPlaceholder base URL
base_url = 'https://jsonplaceholder.typicode.com/posts'

# 2. JSONPlaceholder accepts query parameters like 'userId' or 'id'
payload = {
    'userId': 1,
    'id': 5
}

try:
    # 3. Sending the request
    response = requests.get(base_url, params=payload)
    
    # Check if the request was successful
    response.raise_for_status()

    # 4. Print the final URL and the resulting data
    print(f"Final URL: {response.url}")
    print("Response Data:")
    print(response.json())

except requests.exceptions.ConnectionError:
    print("Connection Error: Check your internet or if the URL is correct.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import requests
import csv

base_Url = 'https://jsonplaceholder.typicode.com/users'

response = requests.get(base_Url)

if response.status_code == 200:
    data = response.json()
    
    with open('user_data.csv', 'w') as file:
        writer = csv.writer(file)
        
        writer.writerow(["ID", "Name", "Email", "City", "Lat", "Long"])
        
        for user in data:
            row = [
                user['id'],
                user['name'],
                user['email'],
                user['address']['city'],
                user['address']['geo']['lat'],
                user['address']['geo']['lng']
            ]
            writer.writerow(row)
    print('Data printed successfully to csv file')    
```

```python
import requests

base_Url = 'https://jsonplaceholder.typicode.com/posts'

search_term = {
    "userId": 5,
    "completed": "false"
}

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36...",
    "Accept": "application/json"
}

response = requests.get(base_Url, params=search_term, headers=headers)

if response.status_code == 200:
    data = response.json()
    
print(data)    
```

```python
import requests

base_Url = 'https://jsonplaceholder.typicode.com/posts'

new_payload = {
    "title": "Mastering python APIs",
    "body": "Get a detailed guide on how to go about AP{Is in Python",
    "userId": 99
}

response = requests.post(base_Url, json=new_payload)

if response.status_code == 201:
    print("Data posted successfully", response.json()['id'])
```

```python
import requests

base_Url = 'https://jsonplaceholder.typicode.com/posts/1'

updated_payload = {
    'id': 1,
    "title": "Mastering Python Regular Expressions",
    "userId": 1
}

response = requests.put(base_Url, json=updated_payload)

if response.status_code == 200:
    print("Data updated successfully", response.json()['id'])
    
print(response)    
```

```python
import requests
from requests.exceptions import HTTPError, Timeout

base_Url = 'https://jsonplaceholder.typicode.com/posts/1'

try:
    response = requests.delete(base_Url)
    
    if response.status_code == 204:
        print("Resource deleted successfully")
except Timeout:
    print("Server took too long")        
except HTTPError as err:
    print("Unexpected error occured: {err}") # 404/ 500       
print(response)
```
